# 1. 라이브러리 설치

```python
import dlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import tensorflow as tf
```

# 2. 모델 불러오기

```python
# 학습된 모델 불러오기
detector = dlib.get_frontal_face_detector()
sp = dlib.shape_predictor("./models/shape_predictor_68_face_landmarks.dat")
```

# 3. 소스파일 불러오기

```python
img = dlib.load_rgb_image("source/1.jpg")
```

```python
# 화면에 표시
plt.figure(figsize=(16, 10))
plt.imshow(img)
```

```python
imgs = img.copy()

dets = detector(imgs, 1)

print(dets)
```

# 5. 검출데이터 얼굴영역 바운딩박스(Bounding Box) 그리기

```python
fig, ax = plt.subplots(1, figsize=(16, 10))

for det in dets:
    x = det.left()
    y = det.top()
    w = det.width()
    h = det.height()
    rect = patches.Rectangle(
        (x, y),
        w,
        h,
        linewidth = 2,
        edgecolor = 'r',
        facecolor = 'none'
    )
    ax.add_patch(rect) # 네번 돈다.
    
ax.imshow(imgs)
```

# 6. 5개의 점으로 사람의 얼굴(랜드마크)를 인식

```python
fig, ax = plt.subplots(1, figsize=(16, 10))

objects = dlib.full_object_detections()

for det in dets:
    s = sp(img, det)
    print(s.parts())
    objects.append(s)
    
    for point in s.parts():
        # 점 Circle
        circle = patches.Circle(
            (point.x, point.y),
            radius = 2,
            edgecolor = 'r',
            facecolor = 'b'
        )
        ax.add_patch(circle)
        
ax.imshow(imgs)
```

# 7. 얼굴 수평작업(Alignment) 작업 - 전처리 작업중

```python
faces = dlib.get_face_chips(img, objects, size=256, padding=0.3)

fig, axes = plt.subplots(1, len(faces)+1, figsize=(20, 16))
axes[0].imshow(img)

for i, face in enumerate(faces):
    axes[i+1].imshow(face)
```

# 8. 얼굴 분리하는 함수 만들기

```python
def align_faces(img):
    dets = detector(img, 1) # 얼굴감지
    
    objs = dlib.full_object_detections() # 얼굴의 랜드마크 포인트 입력
    
    for det in dets:
        s = sp(img, det)
        objs.append(s)
        
    faces = dlib.get_face_chips(img, objs, size=256, padding=0.3)
    
    return faces
```

```python
# 이미지를 넣고 얼굴가죽을 뜯는 함수 테스트
test_img = dlib.load_rgb_image('./source/2.jpg')
test_faces = align_faces(test_img)
```

```python
# 얼굴 추출 표시
faces = dlib.get_face_chips(img, objects, size=256, padding=0.3)

fig, axes = plt.subplots(1, len(test_faces)+1, figsize=(20, 16))
axes[0].imshow(test_img)

for i, face in enumerate(test_faces):
    axes[i+1].imshow(face)
```

# 9. BeautyGAN Pre-Trained 모델 불러오기

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
saver = tf.train.import_meta_graph('./Drive/model.meta')
saver.restore(sess, tf.train.latest_checkpoint('./Drive'))
graph = tf.get_default_graph()
```

```python
X = graph.get_tensor_by_name('X:0') # source
Y = graph.get_tensor_by_name('Y:0') # reference
Xs = graph.get_tensor_by_name('generator/xs:0') # output
# 그냥 이렇게하는거. reference에 나와있다. accustomal
```

# 10. 전처리 함수

```python
# 전처리
def preprocess(img):
    return img.astype(np.float32) / 127.5 - 1.

# 전처리 복원
def postprocess(img):
    return ((img + 1.) * 127.5).astype(np.uint8)
```

# 11. 실제 적용

```python
s_img = dlib.load_rgb_image('./Source/2.jpg')
img1_faces = align_faces(s_img)
```

```python
r_img = dlib.load_rgb_image('./imgs/makeup/XMY-136.png')
img2_faces = align_faces(r_img)
```

```python
# faces = dlib.get_face_chips(img, objects, size=256, padding=0.3)

# fig, axes = plt.subplots(1, len(img1_faces)+1, figsize=(20, 16))
# axes[0].imshow(img1_faces)

# for i, face in enumerate(img1_faces):
#     axes[i+1].imshow(face)
```

# 12. 얼굴 화장 적용하기

```python
src_img = img1_faces[1]
ref_img = img2_faces[0]

# 원본이미지 전처리
X_img = preprocess(src_img)
X_img = np.expand_dims(X_img, axis=0)

# 레퍼런스 이미지 전처리
Y_img = preprocess(ref_img)
Y_img = np.expand_dims(Y_img, axis=0)
```

```python
# BeautyGAN 동작
output = sess.run(Xs, feed_dict={
    X : X_img,
    Y : Y_img
})
```

```python
# 전처리 복원
output_img = postprocess(output[0])
```

```python
# 결과 화면에 표시
fig, axes = plt.subplots(1, 3, figsize=(20, 10))
axes[0].set_title('Source')
axes[0].imshow(src_img)
axes[1].set_title('Reference')
axes[1].imshow(ref_img)
axes[2].set_title('Output')
axes[2].imshow(output_img)
```
